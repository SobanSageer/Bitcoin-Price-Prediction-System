```python
class crypto(object):
    def scrape_data(self):
        pass
    def clean_data(self):
        pass
    def handle_missing_values(self):
        pass
    def scrape_crypto_prices(crypto_symbol):
        pass
    def calculate_log_returns(self):
        pass
    def save_to_csv(self):
        pass
    def display_data(self):
        pass
    def perform_analysis(self):
        pass
    def create_visualizations(self):
        pass
    def plot_historical_prices(self):
        pass
    def plot_mean_returns(self):
        pass
class bitcoin(crypto):
    def __init__(self):
        self.data = self.scrape_data()
    def scrape_data(self):
        url = "https://data.messari.io/api/v1/assets/bitcoin/metrics/price/time-series"
        try:
            response = requests.get(url, verify=True)
            response.raise_for_status()
            data = response.json()['data']['values']
            return pd.DataFrame(data)
        except requests.exceptions.RequestException as e:
            print("Error fetching data for bitcoin: {}".format(e))
            return None
    def clean_data(self):
        self.data.columns = ['time', 'opening_price', 'closing_price', 'high', 'low', 'volume']
        self.data = self.data.dropna()
    def handle_missing_values(self):
        for column in self.data.columns:
            self.data[column] = self.data[column].fillna(self.data[column].mean())
    def calculate_log_returns(self):
        price_columns = ['opening_price', 'closing_price', 'high', 'low']
        for col in price_columns:
            self.data[f'{col}_log_return'] = np.log(self.data[col] / self.data[col].shift(1))

    def save_to_csv(self):
        if self.data is not None:
            csv_filename = "bitcoin_data.csv"
            self.data.to_csv(csv_filename, index=False)
            print("\nData for bitcoin saved to {}".format(csv_filename))
        else:
            print("Failed to save data for bitcoin")

    def display_data(self):
        print("\nData for bitcoin After Handling Missing Values and Calculating Log Returns:")
        print(self.data.head())

    def perform_analysis(self):
        print("\nAnalysis for bitcoin:")
        for col in ['opening_price', 'closing_price', 'high', 'low']:
            mean_return = self.data[col].mean()
            median_return = self.data[col].median()
            std_dev_return = self.data[col].std()

            print("\nAnalysis for {}:".format(col))
            print("Mean: ", mean_return)
            print("Median: ", median_return)
            print("Standard Deviation: ", std_dev_return)

            max_return_date = self.data.loc[self.data[col].idxmax(), 'time']
            max_return_value = self.data[col].max()

            print("\nDate with the Highest Return for {}:".format(col))
            print("Date: {}".format(max_return_date))
            print("Highest Return Value: {}".format(max_return_value))

    def create_visualizations(self):
        self.plot_historical_prices()
        self.plot_mean_returns()
    def plot_historical_prices(self):
        plt.figure(figsize=(9, 6))
        plt.plot(self.data['time'], self.data['closing_price'], label='Closing Price', color='green', marker='o')
        plt.title("Historical Prices of bitcoin")
        plt.xlabel("Date")
        plt.ylabel("Closing Price")
        plt.grid(True)
        plt.savefig("bitcoin_historical_prices.png")
        plt.show()
    def plot_mean_returns(self):
        price_columns = ['opening_price', 'closing_price', 'high', 'low']
        mean_returns = self.data[[col for col in price_columns]].mean()

        plt.figure(figsize=(8, 5))
        mean_returns.plot(kind='bar', color=['yellow', 'indigo', 'beige', 'pink'])
        plt.title("Mean Returns of bitcoin")
        plt.xlabel("Price Type")
        plt.ylabel("Mean Return")
        plt.savefig("bitcoin_mean_returns.png")
        plt.show()
class xrp(crypto):
    def __init__(self):
        self.data = self.scrape_data()
    def scrape_data(self):
        url = "https://data.messari.io/api/v1/assets/xrp/metrics/price/time-series"
        try:
            response = requests.get(url, verify=True)
            response.raise_for_status()
            data = response.json()['data']['values']
            return pd.DataFrame(data)
        except requests.exceptions.RequestException as e:
            print("Error fetching data for xrp: {}".format(e))
            return None
    def clean_data(self):
        self.data.columns = ['time', 'opening_price', 'closing_price', 'high', 'low', 'volume']
        self.data = self.data.dropna()
    def handle_missing_values(self):
        for column in self.data.columns:
            self.data[column] = self.data[column].fillna(self.data[column].mean())
    def calculate_log_returns(self):
        price_columns = ['opening_price', 'closing_price', 'high', 'low']
        for col in price_columns:
            self.data[f'{col}_log_return'] = np.log(self.data[col] / self.data[col].shift(1))

    def save_to_csv(self):
        if self.data is not None:
            csv_filename = "xrp_data.csv"
            self.data.to_csv(csv_filename, index=False)
            print("\nData for xrp saved to {}".format(csv_filename))
        else:
            print("Failed to save data for xrp")

    def display_data(self):
        print("\nData for xrp After Handling Missing Values and Calculating Log Returns:")
        print(self.data.head())

    def perform_analysis(self):
        print("\nAnalysis for xrp:")
        for col in ['opening_price', 'closing_price', 'high', 'low']:
            mean_return = self.data[col].mean()
            median_return = self.data[col].median()
            std_dev_return = self.data[col].std()

            print("\nAnalysis for {}:".format(col))
            print("Mean: ", mean_return)
            print("Median: ", median_return)
            print("Standard Deviation: ", std_dev_return)

            max_return_date = self.data.loc[self.data[col].idxmax(), 'time']
            max_return_value = self.data[col].max()

            print("\nDate with the Highest Return for {}:".format(col))
            print("Date: {}".format(max_return_date))
            print("Highest Return Value: {}".format(max_return_value))

    def create_visualizations(self):
        self.plot_historical_prices()
        self.plot_mean_returns()
    def plot_historical_prices(self):
        plt.figure(figsize=(9, 6))
        plt.plot(self.data['time'], self.data['closing_price'], label='Closing Price', color='green', marker='o')
        plt.title("Historical Prices of xrp")
        plt.xlabel("Date")
        plt.ylabel("Closing Price")
        plt.grid(True)
        plt.savefig("xrp_historical_prices.png")
        plt.show()
    def plot_mean_returns(self):
        price_columns = ['opening_price', 'closing_price', 'high', 'low']
        mean_returns = self.data[[col for col in price_columns]].mean()

        plt.figure(figsize=(8, 5))
        mean_returns.plot(kind='bar', color=['yellow', 'indigo', 'beige', 'pink'])
        plt.title("Mean Returns of xrp")
        plt.xlabel("Price Type")
        plt.ylabel("Mean Return")
        plt.savefig("xrp_mean_returns.png")
        plt.show()
class ethereum(crypto):
    def __init__(self):
        self.data = self.scrape_data()
    def scrape_data(self):
        url = "https://data.messari.io/api/v1/assets/ethereum/metrics/price/time-series"
        try:
            response = requests.get(url, verify=True)
            response.raise_for_status()
            data = response.json()['data']['values']
            return pd.DataFrame(data)
        except requests.exceptions.RequestException as e:
            print("Error fetching data for ethereum: {}".format( e))
            return None
    def clean_data(self):
        self.data.columns = ['time', 'opening_price', 'closing_price', 'high', 'low', 'volume']
        self.data = self.data.dropna()
    def handle_missing_values(self):
        for column in self.data.columns:
            self.data[column] = self.data[column].fillna(self.data[column].mean())
    def scrape_crypto_prices(crypto_symbol):
        pass
    def calculate_log_returns(self):
        price_columns = ['opening_price', 'closing_price', 'high', 'low']
        for col in price_columns:
            self.data[f'{col}_log_return'] = np.log(self.data[col] / self.data[col].shift(1))

    def save_to_csv(self):
        if self.data is not None:
            csv_filename = "ethereum_data.csv"
            self.data.to_csv(csv_filename, index=False)
            print("\nData for ethereum saved to {}".format(csv_filename))
        else:
            print("Failed to save data for ethereum")

    def display_data(self):
        print("\nData for ethereum After Handling Missing Values and Calculating Log Returns:")
        print(self.data.head())

    def perform_analysis(self):
        print("\nAnalysis for ethereum:")
        for col in ['opening_price', 'closing_price', 'high', 'low']:
            mean_return = self.data[col].mean()
            median_return = self.data[col].median()
            std_dev_return = self.data[col].std()

            print("\nAnalysis for {}:".format(col))
            print("Mean: ", mean_return)
            print("Median: ", median_return)
            print("Standard Deviation: ", std_dev_return)

            max_return_date = self.data.loc[self.data[col].idxmax(), 'time']
            max_return_value = self.data[col].max()

            print("\nDate with the Highest Return for {}:".format(col))
            print("Date: {}".format(max_return_date))
            print("Highest Return Value: {}".format(max_return_value))

    def create_visualizations(self):
        self.plot_historical_prices()
        self.plot_mean_returns()
    def plot_historical_prices(self):
        plt.figure(figsize=(9, 6))
        plt.plot(self.data['time'], self.data['closing_price'], label='Closing Price', color='green', marker='o')
        plt.title("Historical Prices of ethereum")
        plt.xlabel("Date")
        plt.ylabel("Closing Price")
        plt.grid(True)
        plt.savefig("ethereum_historical_prices.png")
        plt.show()
    def plot_mean_returns(self):
        price_columns = ['opening_price', 'closing_price', 'high', 'low']
        mean_returns = self.data[[col for col in price_columns]].mean()

        plt.figure(figsize=(8, 5))
        mean_returns.plot(kind='bar', color=['yellow', 'indigo', 'beige', 'pink'])
        plt.title("Mean Returns of ethereum")
        plt.xlabel("Price Type")
        plt.ylabel("Mean Return")
        plt.savefig("ethereum_mean_returns.png")
        plt.show()
    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb

# Load and convert timestamp
df = pd.read_csv('bitcoin_data.csv')
df['time'] = pd.to_datetime(df['time'], unit='ms')

# Create features
df['hour'] = df['time'].dt.hour
df['day_of_week'] = df['time'].dt.dayofweek
df['price_range'] = df['high'] - df['low']
df['volume_ma'] = df['volume'].rolling(window=5).mean()
df['price_ma'] = df['closing_price'].rolling(window=5).mean()
df['prev_close'] = df['closing_price'].shift(1)
df['prev_volume'] = df['volume'].shift(1)

# Prepare features and targets
df = df.dropna()
feature_cols = ['hour', 'day_of_week', 'volume', 'price_range', 
                'volume_ma', 'price_ma', 'prev_close', 'prev_volume']
X = df[feature_cols]
y = df[['opening_price', 'closing_price']]

# Split and scale
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train models
xgb_open = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
xgb_close = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)

xgb_open.fit(X_train_scaled, y_train['opening_price'])
xgb_close.fit(X_train_scaled, y_train['closing_price'])

# Predictions and metrics
open_pred = xgb_open.predict(X_test_scaled)
close_pred = xgb_close.predict(X_test_scaled)

print('Model Performance Metrics:')
print('Opening Price RMSE:', np.sqrt(mean_squared_error(y_test['opening_price'], open_pred)))
print('Opening Price R2:', r2_score(y_test['opening_price'], open_pred))
print('Closing Price RMSE:', np.sqrt(mean_squared_error(y_test['closing_price'], close_pred)))
print('Closing Price R2:', r2_score(y_test['closing_price'], close_pred))

# Visualization of actual vs predicted prices
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
sns.lineplot(x=df['time'].iloc[-len(y_test):], y=y_test['opening_price'], label='Actual Opening Price')
sns.lineplot(x=df['time'].iloc[-len(y_test):], y=open_pred, label='Predicted Opening Price', linestyle='dashed')
plt.legend()
plt.title('Actual vs Predicted Opening Prices')
plt.xlabel('Time')
plt.ylabel('Price')

plt.subplot(2, 1, 2)
sns.lineplot(x=df['time'].iloc[-len(y_test):], y=y_test['closing_price'], label='Actual Closing Price')
sns.lineplot(x=df['time'].iloc[-len(y_test):], y=close_pred, label='Predicted Closing Price', linestyle='dashed')
plt.legend()
plt.title('Actual vs Predicted Closing Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.tight_layout()
plt.show()

# Prepare next day prediction
last_data = df.iloc[-1:]
last_features = scaler.transform(last_data[feature_cols])
next_open = xgb_open.predict(last_features)[0]
next_close = xgb_close.predict(last_features)[0]

print('\nPredictions for next period:')
print('Predicted Opening Price:', next_open)
print('Predicted Closing Price:', next_close)

# Visualization for next predicted prices
plt.figure(figsize=(6, 4))
sns.barplot(x=['Predicted Opening Price', 'Predicted Closing Price'], y=[next_open, next_close], palette='coolwarm')
plt.title('Predicted Prices for Next Period')
plt.ylabel('Price')
plt.show()

# Trading suggestion
plt.figure(figsize=(6, 4))
if next_close > next_open:
    sns.barplot(x=['Suggestion'], y=[1], palette=['green'])
    plt.text(0, 0.5, 'Take Long Position', ha='center', fontsize=12, color='white', weight='bold')
else:
    sns.barplot(x=['Suggestion'], y=[1], palette=['red'])
    plt.text(0, 0.5, 'Take Short Position', ha='center', fontsize=12, color='white', weight='bold')
plt.ylim(0, 1.2)
plt.title('Trading Suggestion')
plt.axis('off')
plt.show()
```
